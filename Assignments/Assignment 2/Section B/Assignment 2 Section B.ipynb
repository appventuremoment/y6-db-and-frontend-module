### Using the database provided, use SQL to answer the queries stated in (a) to (e).

### a)	[2m] 

We define the `quality` for a particular actor as

$\frac{\sum_{\text{all\ movies\ of\ actor}} (\text{num\_rating} \times \text{rating})}{\sum_{\text{all\ movies\ of\ actor}} \text{num\_rating}}$ 

In other words, we define their quality as the weighted average of the ratings of the movies they have played in (regardless of cast position), using the number of ratings for each movie as the weight. 

Find the bottom 15 actors based on the 'quality'. Break ties alphabetically. Display the name and the value of their 'quality'.

```python
%%sql
SELECT name, SUM(num_ratings * rating) / SUM(num_ratings) as quality
FROM movies, play_in
WHERE movies.movieid = play_in.movieid
GROUP BY name
ORDER BY quality ASC, name ASC
LIMIT 15;
```

### b)	[2m] 
Find the names of actors who meet both of the following conditions:
* They have acted in at least 5 movies.
* The average rating of all the movies they have acted in is higher than the overall average rating of all movies in the database.

Sort the results in alphabetical order.

```python
%%sql
SELECT name
FROM movies, play_in
WHERE movies.movieid = play_in.movieid
GROUP BY name
HAVING COUNT(*) >= 5 AND AVG(rating) > (SELECT AVG(rating) FROM movies)
ORDER BY name;
```

### c)	[2m] 

Find the most controversial actor based on the variability of movie ratings they have received. The controversy score of an actor is defined as the maximum difference between the highest and lowest ratings of the movies they have acted in, regardless of their cast position. 

For example, if actor Smith has appeared in movies with ratings ranging from 1.1 to 9.2, their controversy score would be calculated as `9.2  1.1 = 8.1`.

Display the name of the top-most controversial actor, if there is a tie in first place, break it alphabetically. 

```python
%%sql
SELECT name as 'Most controversial actor'
FROM movies, play_in
WHERE movies.movieid = play_in.movieid
GROUP BY name
ORDER BY MAX(rating) - MIN(rating) DESC, name ASC
LIMIT 1;
```

### d)	[2m] 
Retrieve the title, rating, release year, and number of ratings for all movies that meet the following conditions:
* The movie was released in or after the year 2005.
* The number of ratings the movie has received is greater than the average number of ratings of all movies released in or after 2005.

Additionally, for each movie, classify its rating into one of the following categories:
* 'Excellent' if the rating is 8.0 or higher.
* 'Good' if the rating is between 6.5 (inclusive) and 8.0.
* 'Average' if the rating is between 5.0 (inclusive) and 6.5.
* 'Poor' if the rating is below 5.0.

Finally, display the results sorted by the movie's rating in descending order, with the highest-rated movies appearing first.

```python
%%sql
SELECT title, rating, myear as 'Release year', num_ratings as 'Number of ratings',
CASE WHEN rating >= 8 THEN 'Excellent'
WHEN rating >= 6.5 AND rating < 8 THEN 'Good'
WHEN rating >= 5 AND rating < 6.5 THEN 'Average'
WHEN rating < 5 THEN 'Poor' END as "Rating bin"
FROM movies
WHERE myear >= 2005 AND num_ratings > (
    SELECT AVG(num_ratings)
    FROM movies
    WHERE myear >= 2005
)
ORDER BY rating DESC;
```

### e)	[2m] 

Display the count of distinct pairs of movies that have at least two actors in common (ignoring cast position). The count should not include self-pairs and mirror-pairs.

```python
%%sql
SELECT COUNT(*) AS 'Number of distinct pairs'
FROM (
    SELECT p1.movieid AS movie1, p2.movieid AS movie2
    FROM play_in p1, play_in p2 
    WHERE p1.name = p2.name AND p1.movieid < p2.movieid
    GROUP BY p1.movieid, p2.movieid
    HAVING COUNT(DISTINCT p1.name) >= 2
) as temp;
```
