#### b) You are to revise the commands below. Add in relevant commands to specify the primary key, foreign keys and relevant referential integrity constraints (using on delete cascade etc commands) for each table in Company Database.

```python
%%sql
DROP DATABASE IF EXISTS COMPANY;
CREATE DATABASE COMPANY;
USE COMPANY;

CREATE TABLE department (
  dname        varchar(25),
  dnumber      int,
  PRIMARY KEY(dnumber),
  mgrssn       char(9) , 
  mgrstartdate date
);

CREATE TABLE dept_locations (
  dnumber   integer(4) ,
  dlocation varchar(15),
  PRIMARY KEY(dnumber, dlocation),
  FOREIGN KEY(dnumber) REFERENCES DEPARTMENT(dnumber) ON DELETE CASCADE
);

CREATE TABLE project (
  pname      varchar(25) ,
  pnumber    integer(4)  ,
  plocation  varchar(15),
  dnum       integer(4),
  PRIMARY KEY(pnumber, plocation),
  UNIQUE(pnumber),
  FOREIGN KEY(dnum) REFERENCES DEPARTMENT(dnumber) ON DELETE CASCADE
);

CREATE TABLE employee (
  fname    varchar(15) , 
  minit    varchar(1),
  lname    varchar(15) ,
  ssn      char(9)  ,
  bdate    date,
  address  varchar(50),
  sex      char,
  salary   decimal(10,2),
  superssn char(9),
  dno      integer(4),
  PRIMARY KEY(ssn),
  FOREIGN KEY(superssn) REFERENCES EMPLOYEE(ssn) ON DELETE SET NULL
);

CREATE TABLE works_on (
  essn   char(9)  ,
  pno    integer(4) ,
  hours  decimal(4,1),
  PRIMARY KEY(essn, pno),
  FOREIGN KEY(pno) REFERENCES PROJECT(pnumber) ON DELETE CASCADE,
  FOREIGN KEY(essn) REFERENCES EMPLOYEE(ssn) ON DELETE CASCADE
);

CREATE TABLE dependent (
  essn           char(9) ,
  dependent_name varchar(15) ,
  sex            char,
  bdate          date,
  relationship   varchar(8),
  PRIMARY KEY(essn, dependent_name),
  FOREIGN KEY(essn) REFERENCES EMPLOYEE(ssn) ON DELETE CASCADE
);
ALTER TABLE DEPARTMENT ADD FOREIGN KEY (mgrssn) REFERENCES EMPLOYEE(SSN)
```

### Q2 Recipe Database

#### a) Write the SQL DDL to create the following 5 tables describing cooking: Recipe, Ingredient, Author, Cook, Requires:

* An Author table where each author is identified by an integer id and has a name (up to 30 characters).

* An Ingredient table where each ingredient has an id of up to 5 characters and a name (up to 30 characters).

* A Recipe table where each recipe is identified by a field called id that is an integer. Other attributes include name (string up to 40 characters), author id (integer), and directions (string up to 255 characters). Make all foreign keys set to null on delete and cascade on update.

* A Cook table where each time a recipe is made is identified by a date/time (DATETIME). The table also has a recipe id and a comment (string up to 255 characters). Make all foreign keys set to perform no action (generate error) on delete and cascade on update.

* A Requires table that stores what ingredients are needed in a recipe. This table has a recipe id, ingredient id, and amount (floating point number). Make all foreign keys set to cascade on both update and delete.


```python
%%sql
DROP DATABASE IF EXISTS COOKING;
CREATE DATABASE COOKING;
USE COOKING;

CREATE TABLE author (
    authorid int PRIMARY KEY,
    name varchar(30)
);

CREATE TABLE ingredient (
    ingreid varchar(5) PRIMARY KEY,
    name varchar(30)
);

CREATE TABLE recipe(
    recipeid int PRIMARY KEY,
    name varchar(40),
    Aid int,
    directions varchar(255),
    FOREIGN KEY (Aid) REFERENCES AUTHOR(authorid) ON DELETE SET NULL ON UPDATE CASCADE
);

CREATE TABLE cook (
    date datetime,
    Rid int,
    comment varchar(255),
    PRIMARY KEY (date, Rid),
    FOREIGN KEY(Rid) REFERENCES RECIPE(recipeid) ON DELETE NO ACTION ON UPDATE CASCADE
);

CREATE TABLE requires (
    Rid int,
    Ingid varchar(5),
    amount float,
    PRIMARY KEY (Rid, Ingid),
    FOREIGN KEY(Rid) REFERENCES RECIPE(recipeid) ON DELETE CASCADE ON UPDATE CASCADE,
    FOREIGN KEY(Ingid) REFERENCES INGREDIENT(ingreid) ON DELETE CASCADE ON UPDATE CASCADE
);
```
