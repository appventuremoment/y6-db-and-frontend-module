```python
%%sql
DROP DATABASE IF EXISTS COMPANY;
CREATE DATABASE COMPANY;
USE COMPANY;

CREATE TABLE department (
  dname        varchar(25),
  dnumber      int primary key,
  mgrssn       char(9) , 
  mgrstartdate date
);

CREATE TABLE dept_locations (
  dnumber   integer(4) ,
  dlocation varchar(15) ,
    primary key (dnumber, dlocation),
    foreign key (dnumber) references department(dnumber) on delete cascade
);

CREATE TABLE project (
  pname      varchar(25) ,
  pnumber    integer(4)  primary key,
  plocation  varchar(15),
  dnum       integer(4) ,
    foreign key (dnum) references department(dnumber) on delete cascade
);

CREATE TABLE employee (
  fname    varchar(15) , 
  minit    varchar(1),
  lname    varchar(15) ,
  ssn      char(9) primary key ,
  bdate    date,
  address  varchar(50),
  sex      char,
  salary   decimal(10,2),
  superssn char(9),
  dno      integer(4)
);

alter table employee add foreign key (superssn) references employee(ssn) on delete set null;
alter table department add foreign key (mgrssn) references employee(ssn) on delete set null;

CREATE TABLE works_on (
  essn   char(9)  ,
  pno    integer(4) ,
  hours  decimal(4,1),
    primary key(essn, pno),
    foreign key (essn) references employee(ssn) on delete cascade,
    foreign key (pno) references project(pnumber) on delete cascade
);

CREATE TABLE dependent (
  essn           char(9) ,
  dependent_name varchar(15) ,
  sex            char,
  bdate          date,
  relationship   varchar(8),
    primary key (essn, dependent_name),
    foreign key (essn) references employee(ssn) on delete cascade
);
```

#### a) Retrieve department name and manger ssn from table DEPARTMENT

```python
%%sql
SELECT dname, mgrssn 
FROM department;
```

####  b) Retrieve all rows in table PROJECT with PLOCATION=’Stafford’

```python
%%sql
SELECT * 
FROM project 
WHERE plocation='Stafford';
```

#### c) Retrieve the department name and department location for each department location

```python
%%sql
SELECT department.dname, dept_locations.dlocation 
FROM department 
JOIN dept_locations 
ON department.dnumber=dept_locations.dnumber;
```

#### d) Retrieve the name and address of all employees who work for the 'Research' department.

```python
%%sql
SELECT CONCAT(employee.fname, ' ', employee.minit, ' ', employee.lname) as name, employee.address 
FROM employee 
JOIN department 
ON employee.dno = department.dnumber 
WHERE department.dname='Research';
```

#### e) For each employee, retrieve the employee's name, and the name of his or her immediate supervisor.

```python
%%sql
SELECT CONCAT(em.fname, ' ', em.minit, ' ', em.lname) as name, CONCAT(super.fname, ' ', super.minit, ' ', super.lname) as supername 
FROM employee super, employee em 
WHERE super.ssn=em.superssn;
```

#### f) Retrieve all distinct salary values

```python
%%sql
SELECT DISTINCT salary 
FROM employee;
```

#### g) Show the effect of giving all employees who work on the 'ProductX' project a 10% raise.

```python
%%sql
SELECT employee.*, employee.salary*1.1 as increased_salary
FROM employee
JOIN works_on 
ON employee.ssn=works_on.essn 
JOIN project 
ON project.pnumber=works_on.pno 
WHERE pname='ProductX';
```

#### h) Retrieve the last name of employees whose salary is between 70000 (inclusive) and 80000 (inclusive)

```python
%%sql
SELECT lname
FROM employee
WHERE salary>=70000 AND salary<=80000
```

#### i) Retrieve the last name of employees who are born between year 1975 (inclusive) and 1980 (inclusive)

```python
%%sql
SELECT lname
FROM employee
WHERE YEAR(bdate)>=1975 AND YEAR(bdate)<=1980
```

#### j) Retrieve the last name of all employees whose address is in Houston. Here, the value of the ADDRESS attribute must contain the substring 'Houston‘ in it

```python
%%sql
SELECT lname
FROM employee
WHERE address LIKE '%Houston%';
```

#### k) Retrieve the last name of all employees born in 1940s.

```python
%%sql
SELECT lname
FROM employee
WHERE YEAR(bdate)>=1940 AND YEAR(bdate)<1950;
```

#### l) Show the full name of all employees

```python
%%sql
SELECT CONCAT(fname, ' ', minit, ' ', lname) as full_name
FROM employee;
```

#### m) Retrieve the names of all employees who do not have supervisors.

```python
%%sql
SELECT CONCAT(fname, ' ', minit, ' ', lname) as full_name
FROM employee
WHERE superssn IS NULL;
```

#### n) Show the dependents who are child of an employee.

```python
%%sql
SELECT *
FROM dependent
WHERE relationship='Son' OR relationship='Daughter';
```

#### o) Retrieve the name and birthdate of all employees born in 1950s, sorted from the youngest to the oldest.

```python
%%sql
SELECT CONCAT(fname, ' ', minit, ' ', lname) as name, bdate
FROM employee
WHERE YEAR(bdate)>=1950 AND YEAR(bdate)<1960
ORDER BY bdate DESC;
```

#### p) Using the UNION operator, list the salary of employees whose supervisor manages the Headquarters or Software department.

```python
%%sql
SELECT salary
FROM employee
JOIN department
ON department.mgrssn=employee.superssn
WHERE dname='Headquarters'
UNION
SELECT salary
FROM employee
JOIN department
ON department.mgrssn=employee.superssn
WHERE dname='Software';
```
