Admin work is usually messy, and needing to handle the details of hundreds of projects can cause many issues for teachers, which can results in changes made being lost in the process or not reflected in the most recently sent email.
```python
%%sql
DROP DATABASE if EXISTS arp_projects;
CREATE DATABASE arp_projects;
USE arp_projects;

CREATE TABLE external_company (
  company_name varchar(50),
  industry varchar(25),
  contact_email varchar(50),
  PRIMARY KEY(company_name),
  CONSTRAINT CHECK (contact_email LIKE '%@%')
);

CREATE TABLE overseeing_mentor (
  email varchar(50),
  honorific varchar(3),
  surname varchar(25),
  expertise varchar(25),
  PRIMARY KEY(email),
  CONSTRAINT CHECK (honorific IN ('Mr.', 'Ms.', 'Mrs', 'Dr.')),
  CONSTRAINT CHECK (email LIKE '%@%')
);

CREATE TABLE inschool_mentor (
  email varchar(50),
  honorific varchar(3),
  surname varchar(25),
  subject_code char(2),
  PRIMARY KEY(email),
  CONSTRAINT CHECK (email LIKE '%@nus.edu.sg'),
  CONSTRAINT CHECK (subject_code in ('BL', 'MA', 'PC', 'CM', 'EL', 'CS'))
);

CREATE TABLE external_mentor (
  email varchar(50),
  honorific varchar(3),
  surname varchar(25),
  company_name varchar(50),
  FOREIGN KEY(company_name) REFERENCES external_company(company_name) ON DELETE CASCADE,
  PRIMARY KEY(email),
  CONSTRAINT CHECK (email LIKE '%@%')
);

CREATE TABLE student (
  studentid char(8),
  fname varchar(25),
  lname varchar(25),
  year_of_study int(1),
  PRIMARY KEY(studentid),
  CONSTRAINT CHECK (studentid LIKE 'h%'),
  CONSTRAINT CHECK (year_of_study BETWEEN 5 AND 6)
);

CREATE TABLE project (
  internal_code char(5),
  title varchar(100),
  field_of_study varchar(25),
  taken boolean,
  present_ready boolean,
  poster_received boolean,
  OMemail varchar(50) NOT NULL,
  ISMemail varchar(50),
  ecompany_name varchar(50),
  PRIMARY KEY(internal_code),
  FOREIGN KEY(OMemail) REFERENCES overseeing_mentor(email) ON DELETE CASCADE,
  FOREIGN KEY(ISMemail) REFERENCES inschool_mentor(email) ON DELETE SET NULL,
  FOREIGN KEY(ecompany_name) REFERENCES external_company(company_name) ON DELETE CASCADE
);

CREATE TABLE student_project (
  studentid char(8),
  internal_code char(5),
  PRIMARY KEY(studentid, internal_code),
  CONSTRAINT CHECK (studentid LIKE 'h%'),
  FOREIGN KEY(studentid) REFERENCES student(studentid) ON DELETE CASCADE,
  FOREIGN KEY(internal_code) REFERENCES project(internal_code) ON DELETE CASCADE
);

CREATE TABLE publication (
  link varchar(200),
  journal varchar(100),
  publisher varchar(50),
  publication_date date,
  pid char(5),
  PRIMARY KEY(link),
  UNIQUE (pid),
  FOREIGN KEY(pid) REFERENCES project(internal_code) ON DELETE CASCADE
);

CREATE TABLE ssef_project (
  ssef_code char(7),
  forms_received boolean,
  poster_received boolean,
  result varchar(13),
  pid char(5),
  PRIMARY KEY(ssef_code),
  UNIQUE (pid),
  CONSTRAINT CHECK (result IN ('gold', 'silver', 'bronze', 'merit', 'finalist', 'participation', 'unknown')),
  FOREIGN KEY(pid) REFERENCES project(internal_code) ON DELETE CASCADE
);
```

#### Admin Query 1

Admins (teachers) should be able to checl the number of projects ssef projects that are from a specific sector that got a specific result to see if projects from external did well in ssef.

```python
%%sql
SELECT count(*) AS 'Result'
FROM ssef_project
WHERE pid IN (
    SELECT internal_code 
    FROM project, external_company 
    WHERE ecompany_name = company_name
    AND industry = 'Cybersecurity')
GROUP BY result
HAVING result = 'finalist'
```

#### Admin Query 2

Admins (teachers) should be able to check for year 6 students that do not have a project that is ready to be presented at research congress since they need to have at least one to graduate.

```python
%%sql
SELECT studentid, CONCAT(fname, ' ', lname) as name
FROM student
WHERE studentid
NOT IN (
    SELECT DISTINCT studentid
    FROM student_project, project
    WHERE student_project.internal_code = project.internal_code
    AND present_ready = True)
AND year_of_study = 6
```

#### User Query 1

Users (students) should be able to check for available projects that are from a specific sector.

```python
%%sql
SELECT *
FROM project, external_company
WHERE ecompany_name = company_name
AND taken = False
AND industry = 'AI'
```

#### User Query 2

Users (students) should be able to check if the teachers have received their poster for ssef.

```python
%%sql
SELECT ssef_code, pid, title, ssef_project.poster_received
FROM ssef_project, project
WHERE pid in (
    SELECT project.internal_code
    FROM project, student_project
    WHERE project.internal_code = student_project.internal_code
    AND studentid = 'h1060')
AND internal_code = pid
AND NOT ssef_project.poster_received
```
